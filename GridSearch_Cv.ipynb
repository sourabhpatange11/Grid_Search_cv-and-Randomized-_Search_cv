```python
import sklearn.datasets
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
#loading the dataset from sklearn
df= sklearn.datasets.load_breast_cancer()
df
```

```python
#loading the data to Dataframe
Dataframe=pd.DataFrame(df.data,columns=df.feature_names)
Dataframe.head()
```

```python
Dataframe.shape
```

```python
#adding target column in dataframe
Dataframe['label']=df.target
Dataframe.head()
```

```python
Dataframe.shape
```

```python
Dataframe['label'].value_counts()
```

1= Benign

2=Malignant

```python
#Checking missing values
Dataframe.isna().sum()
```

Seprating features(x) and target(y)

```python
X=Dataframe.drop(columns='label',axis=1)
y=Dataframe['label']
```

```python
X.shape,y.shape
```

```python
X.head()
```

```python
y.head()
```

```python
X=np.array(X)
y=np.array(y)
```

# Grid Search cv

It is used to determining the best parameters for our model

```python
#loding svc model
model= SVC()
```

```python
#hyperparametrs

parameters={'kernel':['rbf','linear','poly','sigmoid'],
           'C':[1,5,10,20],}
```

```python
#gridsearch
classifier=GridSearchCV(model,parameters,cv=5)
```

```python
#fitting the data to model
classifier.fit(X,y)
```

```python
classifier.cv_results_
```

```python
#best_parametetrs
best_para=classifier.best_params_
best_para
```

```python
#for highest accuracy

highest_acc=classifier.best_score_
highest_acc
```

```python
#loading the results to pandas dataframe

result=pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
grid_search_result=result[['param_C','param_kernel','mean_test_score']]
```

```python
grid_search_result
```

```python
grid_search_result.max()
```

```python
grid_search_result['mean_test_score'].max()
```

Highest_Accuracy=95.2%

Best Parameters={'C': 10, 'kernel': 'linear'}
